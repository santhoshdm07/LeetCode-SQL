```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Make a GET request to the website
url = 'https://dbdb.io/browse'
response = requests.get(url, verify=False)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the anchor tags with class 'card-title'
anchor_tags = soup.find_all('a', class_='card-title')

# Extract the href attribute and text content for each anchor tag
data = []
for tag in anchor_tags:
    href = tag.get('href')
    name = tag.text
    data.append({'name': name, 'URL': href})

# Create a pandas DataFrame from the data
df = pd.DataFrame(data)

# Print the DataFrame
df.head(2)
```

```python
import requests
from bs4 import BeautifulSoup

# Make a GET request to the website
url = 'https://dbdb.io/db/1010data'
response = requests.get(url, verify=False)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the anchor tags with class 'card-text'
anchor_tags = soup.find_all('p', class_='card-text')

# Initialize a list to store the extracted URLs
urls = []

# Iterate through the anchor tags and extract the URLs
for tag in anchor_tags:
    url = tag.find('a')['href'] if tag.find('a') else 'N/A'
    urls.append(url)

# Print the extracted URLs
print(urls)
```

```python
def extract_urls(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    anchor_tags = soup.find_all('p', class_='card-text')
    urls = [tag.find('a')['href'] if tag.find('a') else 'N/A' for tag in anchor_tags]
    return urls

# Append the extracted URLs into a new column 'Extracted_URLs'
df['Extracted_URLs'] = df['URL'].apply(lambda url: extract_urls(requests.get(url, verify=False).content))

# Print the updated DataFrame
df.head()
```
